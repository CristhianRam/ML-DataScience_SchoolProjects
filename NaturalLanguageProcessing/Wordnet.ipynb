## NetworkX

```python
import networkx as nx
grafo = nx.Graph()
grafo.add_edge("a","b")
grafo.add_edge("a","c")
grafo.add_edge("b","c")
```

```python
nx.draw_networkx(grafo)
```

```python
todos_los_synset = wn.synsets("perro", lang="spa")
grafo = nx.Graph()
for synset in todos_los_synset:
  principal = synset.lemma_names("spa")
  print("synset principal:",principal)
  nombres_principal = principal
  hiperonimos = synset.hypernyms()
  for cada_hiper in hiperonimos:
    #print("hiperonimos:",cada_hiper.lemma_names("spa"))
    nombres_hiper = cada_hiper.lemma_names("spa")
    for prin in nombres_principal:
      for hiper in nombres_hiper:
        grafo.add_edge(prin, hiper)

  hiponimos = synset.hyponyms()
  for cada_hiponimo in hiponimos:
    #print("hiponimos:",cada_hiponimo.lemma_names("spa"))
    nombres_hipo = cada_hiponimo.lemma_names("spa")
    for prin in nombres_principal:
      for hipo in nombres_hipo:
        grafo.add_edge(prin, hipo)

nx.draw_networkx(grafo)
```

```python
print("Centralidad de grado")
print(nx.degree_centrality(grafo))
print("Centre Intercentralidad")
print(nx.betweenness_centrality(grafo))
#print(nx.closeness_centrality(grafo))
```
